```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
heart_df=pd.read_csv("/content/heart_disease_data.csv")
```

```python
X=heart_df.iloc[:,0:13]
```

```python
print(X.head())
```

```python
Y=heart_df['target']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X_test.shape)
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
train_predict=model.predict(X_train)
print(train_predict)
```

```python
print(Y_train)
y_train_array=np.array(Y_train)
d=np.array(y_train_array)
print(y_train_array)
print(y_train_array.ndim)
```

```python
print(train_predict.ndim)
print(type(train_predict))
```

```python
train_accuracy=accuracy_score(train_predict,y_train_array)
```

```python
print(train_accuracy)
```

```python
X_test=scaler.transform(X_test)

test_predict=model.predict(X_test)
```

```python
Y_test=np.array(Y_test)
```

```python
test_accuracy=accuracy_score(test_predict,Y_test)
print(test_accuracy)
```
